## Question 4 : Palindrome

A palindrome is a word that when reversed, yields the same word. Such examples include "bob", "noon","rotator","sagas". In fact, multiple word sentences such as "step on no pets" can also be a palindrome. For the sake of simplicity, we will stick to one word problems. The task in this question is to design a function denoted <i>isPalindrome</i> which determines whether a string provided is in fact a palindrome. In the case that it is, the function should return True and otherwise False.

```python
def isPalindrome(mystring):
    if mystring == mystring[::-1]:
        return True
    else:
        return False
    pass
```

## Question 6: Recursion 

In this course we often mentioned the powerful capabilities of python regarding recursion. Though niche, it can bode very useful in certain situations. To this end, we recall the fibonacci numbers. The fibonacci numbers often occur in nature, where the number of petals in a flower are elements of the fibonnaci number sequence.

For the purpose of this question, we start with 1 as the first element, which in turn is also the second element. From there, the 3rd element of the sequence is the sum of the previous two elements. In this case, the 3rd element would be 2 (1+1), the 4th element would be 3 (2+1) and so on.

Your task is to design a function denoted <i>Fibonacci</i> which utilises recursion to determine the n'th fibonacci number. In addition, impose the following conditions.

<ul>
    <li> The function can only take values of n which are greater than 0.</li>
    <li> The function can only take values of type integer</li>
</ul>

Use the <i>raise</i> or <i>assert</i> functions in conjugation with TypeError or ValueError to yield the desired result.

Mathematically the n'th fibonacci number $F_n$ can be expressed as :

\begin{align}
F_n &= F_{n-1} + F_{n-2}, \quad F_1 = 1, F_2 = 1 \quad n = 1,2,...,N
\end{align}

Please also note below the <i>lru_cache</i> method from the functools library. This method is an example of a decorator. Decorators are any callable python objects which are used to modify a function or a class. In this case, the <i>lru_cache</i> method will store previous values of the fibonnaci numbers in cache so that they do not have to be re-calculated later. The argument maxsize denotes the number of elements the cache should store. To this end, this addition will dramatically speed up the computations of the function. You do not need this method for the code to work, it was merely an illustration for the interested reader.

```python
import functools

@functools.lru_cache(maxsize = 100)
def fibonacci(n):
    """ 
    This function calculates the n'th fibonacci number
    
    n: The requested n'th Fibonacci number
    
    """
    if not n>=0:
        raise ValueError('The function can only take values of n which are greater than 0.')
    elif not type(n) is int:
        raise ValueError('The function can only take values of type integer.')
    elif n <= 1:
        return n
    else:
        return (fibonacci(n-1) + fibonacci(n-2))
    pass
```

```python
fibonacci(10)
```

## Question 10: To Loan or not to Loan?

Consider applying for a mortgage to purchase a house. The monthly payments required (A) for a given number of years (t) can be described by the following amortized loan formula.

\begin{align}
    A = P\left(\frac{\frac{r}{n}(1+\frac{r}{n})^{nt}}{(1+\frac{r}{n})^{nt}-1}\right)
\end{align}
where the variables of the amortized loan are described as:
<ul>
  <li>P - The principal amount.</li>
  <li>r - The interest rate per year expressed as a decimal.</li>
  <li>n - The number of periods per year. The default is 12, i.e 12 monthly payments.</li>
  <li>t - The number of years. </li>
</ul>

Start by defining a function <i>amortize</i> which takes those aforementioned variables and returns the monthly payments, i.e the value of A.

```python
P = 1e5 #Principal
r = 0.06 #Interest Rate
n = 12 #Number of Periods
t = 30 #Number of years

def amortize(P,r,n,t):
    """
    This function calculates the monthly payments of an amortized loan.
    """
    A = P*((r/n)*(1+(r/n))**(n*t))/((1+r/n)**(n*t)-1)
    return A
# I just inserted the formula
```

```python
print('Per Month I owe €',round(amortize(P,r,n,t),2))
print('This is €',round(amortize(P,r,n,t)*12,2),'per year')
```

### Completing the Amortized Loan Schedule

Your task is to recreate the entire amortized loan schedule. The functions <i>amortize</i> and <i>interest_only</i> need not be used but are an aid in explaining and helping to solve the question. Please also note that the following:

<ul>
<li>The amortized loan value is fixed from the beginning of the loan period until the end of the loan period.</li>
<li>The interest for each month is not constant, it varies at each iteration, or each month.</li>
<li>Consider you start the loan from today. Furthermore, please make use of the provided libraries. <i>date.today()</i> and <i>td.relativedelta()</i> can be especially important.
</ul>

```python
import pandas as pd
from datetime import date
import numpy as np
import matplotlib.pyplot as plt
import dateutil.relativedelta as td

#example: https://pbpython.com/amortization-model.html

P = 1e5 #Principal
r = 0.06 #Interest Rate
n = 12 #Number of Periods
t = 30 #Number of years

def amortized_schedule(P, r, n,t,start_date = date.today()):
    """
    P : The principal
    r : the interest rate as a fraction
    n : The number of payment periods. Default is 12.
    t : The number of years required to pay the loan.
    """
    # Construct the index with the DateTimeIndex
    start_date=date.today()
    rng = pd.date_range(start_date, periods=t*n, freq='M')+ pd.DateOffset(days=date.today().day)
    rng.name = 'Start Date'
    
    # Do it similar for End Date
    end_date=date.today()+ td.relativedelta(months=1)
    rng2 = pd.date_range(end_date, periods=t*n, freq='M')+ pd.DateOffset(end_date.day)


    #First construct a dataframe:
    df = pd.DataFrame(index=rng,columns=['Period', 'Start Balance', 'Monthly Payment', 'Principal', 'Interest', 'End Balance', 'End Date' ], dtype='float')

    #Set the index to a running number instead of a date
    df.reset_index(inplace=True)

    #Add the End Date, which is exactly one month later than the Start Date
    df['End Date']=rng2

    # Add the period
    df['Period']=df.index+1

    # Now fill in the columns

    #First: monthly payments with the formula from amortize() -> could also use that function
    
    df['Monthly Payment']=P*((r/n)*(1+(r/n))**(n*t))/((1+r/n)**(n*t)-1)

    # Fill in first row wih the Start Balance
    df['Start Balance']=P
    df['Interest']=(r/n)*P
    df['Principal']= df['Monthly Payment'] - df['Interest']
    df['End Balance']= df['Start Balance']-df['Principal']

    #Then loop through the others to fill the other rows
    for i in range (1,n*t):
        df.at[i, 'Start Balance']=df.at[i-1, 'End Balance']
        df.at[i,'Interest']= (r/n)*df.at[i,'Start Balance']
        df.at[i,'Principal']= df.at[i, 'Monthly Payment'] - df.at[i,'Interest']
        df.at[i,'End Balance']= df.at[i,'Start Balance']-df.at[i,'Principal']

    # Round to two decimal places
    df = df.round(2)


    return df
    pass  

schedule1 = amortized_schedule(P = 1e5,r= 0.06,n=12,t=30)
schedule1.head()
```

# Question 12: My First Calculator
Design a class denoted Calculator:

The following functions should be included
- Addition: Computing the sum of N numbers (N can be any amount of numbers) ex : calculator.add(1,2,3,4,5,6) = 21
- Subtraction : Subtract two numbers, Example: calculator:sub(10,5)=5
- Power : Compute the given power of a number: calculator.power(10,2)=100
- Division: Divide two numbers a,b. ex: calculator.divide(10,2)=5 
        - Note: Dividing by zero should not be allowed and thus print "Cannot divide by zero". calculator.divide(10,0)='Cannot divide by zero)'
- Multiply: multiply N numbers together to get the final product. You are not allowed to use any package for this (hence numpy math etc). Ex: calculator.multiply(5,6,7,3)=630

- save: This function can store up to N numbers in the calculator (Note this should not be a variable)
    - Ex: calculator.save(3,4,5,6,7) will store these numbers to be called later. So you can use: calculator.saved= [3,4,5,6,7]
    
Hint: Create all the separate methods as functions first and then convert the set of functions to a class.

```python
#Create you calculator class here
class Calculator(object):
    def __init__(self,a,b):
        self.a = a
        self.b = b
        self.saved = []
    
    def add(self, *args):
        print(args)
        sums= sum(args) + self.a + self.b
        return sums
    
    def subtract(self):
        return self.a - self.b
    
    def multiply(self):
        return self.a * self.b
    
    def divide(self):
        if self.b==0: 
            return "Cannot divide by zero" 
        else: 
            return self.a / self.b
    
    def save(self, *args):
        self.saved=args
        return self.saved
    pass
```
